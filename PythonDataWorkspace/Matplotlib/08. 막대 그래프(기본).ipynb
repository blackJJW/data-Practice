# 8. 막대 그래프(기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용시, 마이너스 글자가 깨지는 현상 해결
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]

plt.bar(labels, values)
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]

plt.bar(labels, values, color='r')
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]
colors = ['r', 'g', 'b']
plt.bar(labels, values, color=colors)
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]
colors = ['r', 'g', 'b']
plt.bar(labels, values, color=colors, alpha=0.5)
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]

plt.bar(labels, values)
plt.ylim(175, 195)
```

```python


plt.bar(labels, values, width=0.5) # 막대가 얇아짐
```

```python
plt.bar(labels, values, width=0.3)
plt.xticks(rotation=45) # x 축의 이름 데이터 각도를 45 도로 설정
plt.yticks(rotation=45) # y 축의 키 데이터 각도를 45 도로 설정
```

```python
ticks = ['1번학생', '2번학생', '3번학생']
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]

plt.bar(labels, values)
plt.xticks(labels, ticks)
```

```python
plt.bar(labels, values)
plt.xticks(labels, ticks, rotation=90)
```
