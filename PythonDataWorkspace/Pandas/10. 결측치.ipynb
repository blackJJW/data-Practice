## 데이터 제외하기 dropna

```python
df = pd.read_csv('score.csv')
df
```

```python
df.dropna() # NaN 을 포함하는 데이터 삭제
```

```python
df
```

- axis : index or columns
- how : any or all

```python
df.dropna(axis = 'index', how='any') # NaN 가 하나라도 있는 row 삭제
```

```python
df.dropna(axis='columns') # NaN이 하나라도 있는 column 삭제
```

```python
df['학교'] = np.nan
df
```

```python
df.dropna(axis='columns', how='all') # column 데이터 전체가 NaN인 경우에만 Column 삭제
```
