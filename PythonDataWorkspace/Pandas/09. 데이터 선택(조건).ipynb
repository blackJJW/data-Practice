# 9. 데이터 선택(조건)
조건에 해당하는 데이터 선택

```python
import pandas as pd
df = pd.read_csv('score.csv')
df
```

```python
df['키'] >= 185
```

```python
filt = (df['키'] >= 185)
df[filt]
```

```python
df[~filt] # ~ : filter를 역으로 적용
```

```python
df[df['키'] >= 185]
```

```python
df.loc[df['키']>=185, '수학'] # 카가 185이상인 학생들의 수학 데이터
```

```python
df.loc[df['키'] >= 185, ['이름', '수학', '과학']] # 키가 185이상인 학샏들의 이름, 수학, 과학 데이터
```

## 다양한 조건


### &

```python
df.loc[(df['키'] >= 185) & (df['학교'] == '북산고')]
```

### | 

```python
df.loc[(df['키'] < 170) | (df['키'] > 200)]
```

### str 함수

```python
filt = df['이름'].str.startswith('송')
df[filt]
```

```python
filt = df['이름'].str.contains('태')
df[filt]
```

```python
df[~filt]
```

```python
langs = ["Python", "Java"]
filt = df["SW특기"].isin(langs) # SW특기가 Python이거나 Java인 사람
```

```python
df[filt]
```

```python
langs = ["python", "java"]
filt = df['SW특기'].str.lower().isin(langs)
df[filt]
```

```python
filt = df['SW특기'].str.contains('Java', na=True) # NaN 데이터에 대해 True로 설정
df[filt]
```

```python
filt = df['SW특기'].str.contains('Java', na=False) # NaN 데이터에 대해 True로 설정
df[filt]
```
