```python
import pandas as pd
df_m = pd.read_excel('인구피라미드.xlsx', skiprows=3, index_col='행정기관', usecols='B, E:Y')
df_m.head()
```

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(',', '').astype(int) # 1,195,951 -> 1195951 (정수형)
```

```python
df_w = pd.read_excel('인구피라미드.xlsx', skiprows=3, index_col='행정기관', usecols='B, AB:AV')
df_w.head()
```

```python
df_m.columns
```

```python
df_w.columns
```

```python
df_w.columns = df_m.columns
df_w.columns
```

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(',', '').astype(int) # 1,195,951 -> 1195951 (정수형)
df_w.iloc[0]
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
plt.figure(figsize=(10, 7))
plt.barh(df_m.columns, -df_m.iloc[0] // 1000)
plt.barh(df_w.columns, df_w.iloc[0] // 1000)
plt.title('2011 대한민국 인구 피라미드')
plt.savefig('2011_인구피라미드.png', dpi=100)
plt.show()
```
